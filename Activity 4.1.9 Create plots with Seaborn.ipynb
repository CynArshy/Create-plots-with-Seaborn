```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")

# view the DataFrame
print(movies.shape)
print(movies.dtypes)
print(movies.columns)
movies.head()
```

```python
ott = pd.read_csv("ott.csv")

# view the DataFrame
print(ott.shape)
print(ott.dtypes)
print(ott.columns)
ott.head()
```

```python
movies_ott = pd.concat([movies, ott]) 

movies_ott.shape
```

```python
# create a countplot to determine what age group has streamed the most movies
sns.countplot(x="Age", data=movies_ott)
```

a)The age group had the most-streamed movies are 18+ 

b)There's a big gap between 16+ and 18+ in movie in contents for the gap

```python
# create the histogram
sns.histplot(data=movies_ott, x="IMDb")
```

```python
sns.histplot(data=movies_ott, x="IMDb",binwidth=1)
```

From the first histogram we can identify some outliers whereas from the second one we can't

```python
# create the scatterplot
sns.scatterplot(x="Rotten Tomatoes", y="IMDb", data=movies_ott)
```

a)What can you infer about the correlation? There's a negative correlation and it is overplotted.

b)If there is a correlation, is the relationship between the ratings strong or weak and positive or negative? there's a strong correlation and it is negative
